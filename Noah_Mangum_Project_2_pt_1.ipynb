```python
import pandas as pd
import numpy as np
```

```python
from scipy import stats
```

```python
pearson_coef, p_value = stats.pearsonr(df['CMP'], df['YDS'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)  
```
